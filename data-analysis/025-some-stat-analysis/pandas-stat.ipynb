```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv(
    'data/sales.csv', 
    parse_dates=['date'], 
    index_col=['date'],
)
```

```python
df
```

## Find the record with the biggest difference

```python
df['diff'] = df['sales'] - df.shift(1)['sales'] 
df['diff_abs'] = df['diff'].abs()
max_row = df['diff_abs'].idxmax()
df.loc[max_row].to_frame().T
```

## Find the record with the biggest difference from multiple categories

```python
df = pd.read_csv('data/sales_multi_products.csv')
df
```

```python
def get_max_diff_record(x_df):
    x_df['diff'] = x_df['quantity'] - x_df.shift(1)['quantity'] 
    x_df['diff_abs'] = x_df['diff'].abs()
    max_row = x_df['diff_abs'].idxmax()
    return x_df.loc[max_row].to_frame().T

df.groupby('product').apply(get_max_diff_record)
```

## Get top 3 diff records from each product

```python
def get_top_3_diff_records(x_df):
    x_df['diff'] = x_df['quantity'] - x_df.shift(1)['quantity'] 
    x_df['diff_abs'] = x_df['diff'].abs()
    
    x_df['rank'] = x_df['diff_abs'].rank(ascending=False)
    cond = x_df['rank'] <=3
    return x_df[cond]

df.groupby('product').apply(get_top_3_diff_records)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Simple plot with 4 numbers

```python
plt.plot([1,3,2,4])
plt.show()
```

## 2. Points have x and y values, add title and axis label

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.title('This is a title', fontsize=8, color='g')
plt.xlabel('number n')
plt.ylabel('n^2')
plt.show()
```

## 3. Change figure size, plot red dots, set axis scales x: 0-6 and y: 0-20

```python
plt.figure(figsize=(2,5))                  # 2 inches wide X 5 inches tall
plt.plot([1,2,3,4], [1,4,9,16], 'ro')      # red o
plt.axis([0, 6, 0, 20])                    # [xmin, xmax, ymin, ymax]
plt.annotate('square it', (3,6))
plt.show()
```

## 4. Bar chart with four bars

```python
plt.clf()
x = np.arange(4)
y = [8.8, 5.2, 3.6, 5.9]
plt.xticks(x, ('A', 'B', 'C', 'D'))
plt.bar(x, y)
# plt.bar(x, y, color='y')
# plt.bar(x, y, color=['lime', 'r', 'k', 'tan'])
plt.show()
```

## 5. Two sets of 10 random dots plotted

```python
d = {
    'Red O': np.random.rand(10),
    'Grn X': np.random.rand(10)
}
df = pd.DataFrame(d)
df.plot(style=['ro', 'gx'])
plt.show()
```

## 6. Time series - six months of random floats

```python
ts = pd.Series(np.random.randn(180), index = pd.date_range('1/1/2018', periods=180))
df = pd.DataFrame(np.random.randn(180, 3), index=ts.index, columns=list('ABC'))
df.cumsum().plot()
plt.show()
```

## 7. Random dots in a scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30 * np.random.rand(N)) ** 2     # 0 to 15 point radii
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.show()
```

## 9.1 Subplots part 1

```python
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1, 10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

## 9.2 Subplots part 2

```python
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')
x = range(0,200, 5)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color='r')
plots[1].scatter(x, y)
plt.show()
```

## 10. Save figure to image file

```python
plt.figure(figsize=(4,3), dpi=100)
plt.plot([9,8,7,6,5,4,3,2,1])
plt.axis([0,7, 0, 10])
plt.title('Flight data')
plt.xlabel('Speed')
plt.savefig('Flight.png')
plt.show()
```
